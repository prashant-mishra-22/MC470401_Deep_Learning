**Experiment 3: Implementation and exploration of the gradient descent algorithm for minimizing a cost function**
---


    
---
    We'll minimize a simple quadratic function: f(x) = x^2 + 2x.
---
    Its derivative (gradient) is f'(x) = 2x + 2.
---
    This quadratic has a minimum at x = -1 (where f(-1) = -1 )
---
    Cost Function and Gradient: f(x) = x^2 + 2x, Grad=2x +2



---

```python
#define cost function
def f(x):
    return (x*x)+(2*x)
```

```python
#define gradient
def df(x):
    return (2*x)+2
```

```python
global_min=-1.0
```

```python
print("Minimum value of f(X) = ",f(global_min)," is at = ",global_min)
```

---


**Parameters and Execution:**
---


---

    start_x = 3.0: Starts far from the minimum to show convergence.
---
    learning_rate = 0.1: Balanced step size (experiment with 0.01-0.3).
---
    num_iterations = 10: Enough for convergence;
---
    monitor if it stabilizes early.
---
    Prints starting/final values and cost.
---


```python
#gradient descent function
def gradient_descent(weight,learning_rate,iterations):
    # Print results
    print("*"*60)
    print(f"Starting point: {weight[0]}")
    print("*"*60)
    #iterations
    for i in range(iterations+1):
        #initial point
        xcurr=weight[i]
        #current grad
        grad_curr=df(xcurr)
        #learning rule
        xnew=xcurr-learning_rate*grad_curr
        #appending tolist
        weight.append(xnew)
        #printing result
        print(f"[Iteration] : {i} ==> [X{i}] : {xcurr} ==> [X{i+1}] : {xnew}")
    print("*"*60)
    print(f"Final x: {weight[len(weight)-1]} (Expected: -1.0)")
    print("-"*80)
    print(f"Final cost: {f(weight[len(weight)-1])}")
    print("*"*60)
    return weight
```

```python
#parameters
x1=[3.0] #starting point
x2=[3.0]
x3=[3.0]
n1=0.01
n2=0.1
n3=0.3
iter=10
```

```python
x1=gradient_descent(x1,n1,iter)
```

```python
x2=gradient_descent(x2,n2,iter)
```

```python

x3=gradient_descent(x3,n3,iter)
```

```python
# Compute cost values for each learning rate
cost1 = [f(x) for x in x1]
cost2 = [f(x) for x in x2]
cost3 = [f(x) for x in x3]

# Iteration indices
iters1 = [i for i in range(0,iter+2)]
iters2 = [i for i in range(0,iter+2)]
iters3 = [i for i in range(0,iter+2)]

# Plot iteration vs cost
plt.figure(figsize=(10,5))

plt.plot(iters1, cost1, label=f"Learning Rate = {n1}", marker='o')
plt.plot(iters2, cost2, label=f"Learning Rate = {n2}", marker='s')
plt.plot(iters3, cost3, label=f"Learning Rate = {n3}", marker='^')

plt.title("Iteration vs Cost for Different Learning Rates   |   Prashant_Kumar_Mishra")
plt.xlabel("Iteration")
plt.ylabel("Cost: f(x)=x^2+2x")
plt.legend()
plt.grid(True)
plt.show()
```
